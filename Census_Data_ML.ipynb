
***1. Replace ? to null for the whole dataset***

```
df_adult.replace('?', np.nan, inplace=True)
df_adult['income'] = df_adult['income'].str.replace('.', '')
```

***2. Trim leading spaces***

```
df_adult = df_adult.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

***3. Fill Occupation and Workclass based on common Income groups. Fill Native country based on most common race for that country***

```
common_Occ = df_adult.groupby(['income','native-country'])['occupation'].agg(lambda x: x.mode()[0]).reset_index(name='comm_occupation')
common_wc= df_adult.groupby(['occupation'])['workclass'].agg(lambda x: x.mode()[0]).reset_index(name='comm_workclass')
common_native= df_adult.groupby(['race'])['native-country'].agg(lambda x: x.mode()[0]).reset_index(name='comm_native')
```

```
common_Occ_dict = dict(zip(zip(common_Occ['native-country'], common_Occ['income']), common_Occ['comm_occupation']))
common_wc_dict=dict(zip(common_wc['occupation'], common_wc['comm_workclass']))
common_native_dict=dict(zip(common_native['race'], common_native['comm_native']))
```

```
df_adult['native-country']=df_adult.apply(lambda row: common_native_dict[row['race']] if pd.isna(row['native-country']) else row['native-country'], axis=1)
```

```
df_adult['occupation'] = df_adult.apply(lambda row: common_Occ_dict[(row['native-country'], row['income'])] if pd.isna(row['occupation']) else row['occupation'], axis=1)
```

```
df_adult['workclass'] = df_adult.apply(lambda row: common_wc_dict[row['occupation']] if pd.isna(row['workclass']) else row['workclass'], axis=1)
```

***4. Change  1-12th as 'Did not graduate' and drop column Education Number***

```
df_adult.head(3)
```

```
df_adult['education'] = df_adult['education'].apply(lambda x: 'School' if x == '11th' or x == '7th-8th' or x == '10th'
                                              or x == '5th-6th' or x == '9th' or x == '12th' or x == '1st-4th'
                                              or x == 'Preschool' else x)
df_adult['education'] = df_adult['education'].apply(lambda x: 'Associate' if x == 'Assoc-acdm' or x == 'Assoc-voc' else x)
education_map = {'School':1,
             'HS-grad':2,
             'Some-college':3,
             'Bachelors':4,
             'Prof-school':5,
             'Associate':6,
             'Masters':7,
             'Doctorate':8}
df_adult['education'] = df_adult['education'].map(education_map)
```

```
df_adult.drop(columns=['education-num'], inplace=True)
```

***5. Add column CG_Category 0: No capital gains, 1: capital gains, 2: Undeclared capital gains***

```
df_adult['CG_Category'] = np.where(df_adult['capital-gain']==0, 0,
                   np.where(df_adult['capital-gain']==99999 ,2,1))
```

```
X = df_adult_new.drop(['income'], axis=1)

y = df_adult_new['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```
encoder = ce.OrdinalEncoder(cols=['workclass', 'education', 'marital-status', 'relationship', 'race', 'sex','native-country','occupation'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier , X=X_train , y=y_train , cv=10)
print('Accuracy: {:.2f} %'.format(accuracies.mean()*100))
print('Standard deviation: {:.2f} %'.format(accuracies.std()*100))
```

```
from sklearn.model_selection import RandomizedSearchCV
param_grid = {"C":np.logspace(-3,3,7), "penalty":["l1","l2"]}
grid = RandomizedSearchCV(classifier, param_grid,cv=10, random_state = 0)
grid.fit(X_train, y_train)
```

```
print(grid.best_params_)
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```
